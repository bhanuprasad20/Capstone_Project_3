```python
# Heatmap
```

```python
plt.figure(figsize = (20,30))
sns.heatmap(corre, cmap = "coolwarm", annot = True)
plt.title("Heat Map of Cars", fontsize = 24)
plt.show()
```

```python
# Pairplot
```

```python
plt.figure(figsize = (20,10))
sns.pairplot(df_clean2)
plt.show()
```

```python
# Lineplot of engine capactiy and power
```

```python
plt.figure(figsize = (20,10))
plt.grid()
plt.plot(df_clean2["engine_displacement"], color = "c", marker = "o", linestyle = "dashed", linewidth = 2, markersize = 10, label = "Engine Capacity (in CC)")
plt.plot(df_clean2["max_power_bhp"], color = "m", marker = ".", linestyle = ":", linewidth = 2, markersize = 10, label = "Power(in BHP)")
plt.xlabel("Engine Capacity (in CC)", fontsize = 16)
plt.ylabel("Power (in BHP)", fontsize = 16)
plt.title("Engine Capacity and Power", fontsize = 24)
plt.legend(["Engine Capacity","Power"], loc = "best")
plt.show()
```

```python
ser_cc = pd.Series(df["engine_displacement"][:30])
print(ser_cc)
print(ser_cc.max())
```

```python
ser_power = pd.Series(df["max_power_bhp"][:30])
print(ser_power)
print(ser_power.max())
```

```python
# Barplot of engine capacity and power
```

```python
plt.figure(figsize = (20,10))
plt.grid()
plt.bar(ser_power, 0, width = 10, color = "c", label = "Engine Capacity")
plt.bar(ser_cc, 1997, width = 10, color = "m", label = "Power")
plt.title("Engine Engine Capacity and Power")
plt.xlabel("Engine Capacity")
plt.ylabel("Power (in BHP)")
plt.legend(["Engine Capacity", "Power"], loc = "best")
plt.show()
```

```python
# Barplot of engine capacity and power
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x = "engine_displacement", y = "max_power_bhp", data = df_clean3, color = 'b', 
            palette = "GnBu_r", saturation = 6, errcolor = '0.9')
plt.show()
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x = "engine_displacement", y = "max_power_bhp", data = df_clean3, hue = "body_type", color = 'b', 
            palette = "GnBu_r", saturation = 6, errcolor = '0.9')
plt.show()
```

```python
# Scatterplot of engine capacity and power
```

```python
plt.figure(figsize = (20,10))
sns.scatterplot(x = "engine_displacement", y = "max_power_bhp", data = df_clean3, hue = "body_type")
plt.show()
```

```python
# Boxplot of engine capacity and power
```

```python
plt.figure(figsize = (20,10))
sns.boxplot(x = "engine_displacement", y = "max_power_bhp", data = df_clean3, hue = "body_type")
plt.show()
```

```python
# Lineplot of Torque and RPM
```

```python
plt.figure(figsize = (20,10))
plt.grid()
plt.plot(df_clean2["max_torque_nm"], color = "c", marker = "o", linestyle = "dashed", linewidth = 2, markersize = 10, label = "Max Torque")
plt.plot(df_clean2["max_torque_rpm"], color = "m", marker = ".", linestyle = ":", linewidth = 2, markersize = 10, label = "Max RPM")
plt.xlabel("Max Torque", fontsize = 16)
plt.ylabel("Max RPM", fontsize = 16)
plt.title("Torque and RPM", fontsize = 24)
plt.legend(["Torque","RPM"], loc = "best")
plt.show()
```
